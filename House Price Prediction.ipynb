```python
import  numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data  = pd.read_csv("housing.csv")
```

```python
data
```

```python
data.info()
```

```python
data.dropna(inplace = True)
```

```python
data.info()
```

```python
from sklearn.model_selection import train_test_split

X = data.drop(["median_house_value"], axis = 1)
y = data["median_house_value"]
```

```python
X
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
train_data = X_train.join(y_train)
```

```python
train_data
```

```python
train_data.hist(figsize = (15,7))
```

```python
train_data.corr(numeric_only=True)  #it gives a correlation matrix here.
```

```python
plt.figure(figsize = (15,8))
sns.heatmap(train_data.corr(numeric_only=True), annot = True, cmap="YlGnBu")
```

### Preprocessing

```python
train_data["total_rooms"] = np.log(train_data['total_rooms'] + 1)
train_data["total_bedrooms"] = np.log(train_data['total_bedrooms'] + 1)
train_data["population"] = np.log(train_data['population'] + 1)
train_data["households"] = np.log(train_data['households'] + 1)
```

```python
train_data.hist(figsize = (15, 8))
```

```python
train_data.ocean_proximity.value_counts()
```

```python
pd.get_dummies(train_data.ocean_proximity) #convert the value of ocean proximity into 0 & 1
```

```python
train_data = train_data.join(pd.get_dummies(train_data.ocean_proximity)).drop(["ocean_proximity"], axis=1)
```

```python
train_data
```

```python
plt.figure(figsize = (15,8))
sns.heatmap(train_data.corr(numeric_only=True), annot = True, cmap="YlGnBu")
```

```python
plt.figure(figsize = (15,8))
sns.scatterplot(x = "latitude", y="longitude", data=train_data, hue="median_house_value", palette="coolwarm")
```

```python
#number of bedrooms per rooms
train_data['bedroom_ratio'] = train_data['total_bedrooms']/train_data['total_rooms']
train_data["household_rooms"] = train_data['total_rooms']/train_data['households']       #rooms per household
```

```python
plt.figure(figsize = (15,8))
sns.heatmap(train_data.corr(numeric_only=True), annot = True, cmap="YlGnBu")
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train, y_train = train_data.drop(['median_house_value'], axis=1), train_data['median_house_value']
X_train_s = scaler.fit_transform(X_train)

reg = LinearRegression()
reg.fit(X_train_s, y_train)
```

```python
test_data = X_test.join(y_test)

test_data["total_rooms"] = np.log(test_data['total_rooms'] + 1)
test_data["total_bedrooms"] = np.log(test_data['total_bedrooms'] + 1)
test_data["population"] = np.log(test_data['population'] + 1)
test_data["households"] = np.log(test_data['households'] + 1)

test_data = test_data.join(pd.get_dummies(test_data.ocean_proximity)).drop(["ocean_proximity"], axis=1)

#number of bedrooms per rooms
test_data['bedroom_ratio'] = test_data['total_bedrooms']/test_data['total_rooms']
test_data["household_rooms"] = test_data['total_rooms']/test_data['households']       #rooms per household
```

```python
X_test, y_test = test_data.drop(['median_house_value'], axis=1), test_data['median_house_value']
```

```python
X_test_s = scaler.transform(X_test)
```

```python
reg.score(X_test_s, y_test) #Accuracy of Linear Regression model
```

## Random Forest

```python
from sklearn.ensemble import RandomForestRegressor

forest = RandomForestRegressor()

forest.fit(X_train_s, y_train)
```

```python
forest.score(X_test_s, y_test) #Accuracy of Random Forest Regression
```

```python
from sklearn.model_selection import GridSearchCV


param_grid = {
    "n_estimators":[100, 200, 300],
    "min_samples_split": [2, 4],
    "max_depth": [4, 8]
}

grid_search = GridSearchCV(forest, param_grid, cv=5,
                          scoring="neg_mean_squared_error",
                          return_train_score=True)
grid_search.fit(X_train_s, y_train)
```

```python
forest = grid_search.best_estimator_
```

```python
forest.score(X_test_s, y_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error

# Predict using the linear regression model
y_pred_lr = reg.predict(X_test_s)

# Calculate MAE and RMSE
mae_lr = mean_absolute_error(y_test, y_pred_lr)
rmse_lr = np.sqrt(mean_squared_error(y_test, y_pred_lr))

print("Linear Regression:")
print("MAE:", mae_lr)
print("RMSE:", rmse_lr)
```

```python
# Predict using the random forest model
y_pred_rf = forest.predict(X_test_s)

# Calculate MAE and RMSE
mae_rf = mean_absolute_error(y_test, y_pred_rf)
rmse_rf = np.sqrt(mean_squared_error(y_test, y_pred_rf))

print("Random Forest:")
print("MAE:", mae_rf)
print("RMSE:", rmse_rf)
```
